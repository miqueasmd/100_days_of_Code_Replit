````python
#Handling ESC Key

# Import required libraries
import random, os, time
from pynput import keyboard  # Library for keyboard event handling
import sys  # For system-related operations

# Global variable to track if user wants to exit
should_exit = False

def on_press(key):
    """Callback function that runs every time a key is pressed"""
    global should_exit  # Access the global variable
    try:
        if key == keyboard.Key.esc:  # Check if ESC key was pressed
            print("\nExiting game...")
            should_exit = True  # Set flag to exit game
    except:
        pass  # Ignore any errors from key processing

# Create and start keyboard listener in background
listener = keyboard.Listener(on_press=on_press)
listener.start()

# Game mechanics functions
def rollDice(side):
    """Generate random number between 1 and specified sides"""
    result = random.randint(1,side)
    return result

def health():
    """Calculate character health using dice rolls"""
    healthStat = ((rollDice(6)*rollDice(12))/2)+10
    return healthStat

def strength():
    """Calculate character strength using dice rolls"""
    strengthStat = ((rollDice(6)*rollDice(8))/2)+12
    return strengthStat

print("Press ESC at any time to exit")

# Main game loop
while not should_exit:  # Continue until ESC is pressed
    print("‚öîÔ∏è CHARACTER BUILDER ‚öîÔ∏è")
    print()
    
    # Name input loop - continues until valid name or ESC
    while not should_exit:
        name = input("Name your Legend:\n")
        if name.strip():  # Check if name isn't just whitespace
            break
        print("Name cannot be empty. Please enter a valid name.")
    
    # Exit check after name input
    if should_exit:
        break
        
    # Get character type
    character_type = input("Character Type (Human, Elf, Wizard, Orc):\n")
    print()
    
    # Display character stats
    print(f"{name} the {character_type}")
    print("HEALTH:", health())
    print("STRENGTH:", strength())
    print()
    print("May your name go down in Legend‚Ä¶")
    print()
    
    # Ask to play again
    again = input("Again? (yes/no):\n").lower()
    if again == "no" or should_exit:  # Exit if 'no' or ESC pressed
        break
    
    # Clear screen and pause briefly between games
    time.sleep(1)
    os.system('cls' if os.name == 'nt' else 'clear')  # Clear screen (works on Windows and Unix)

# Clean up keyboard listener when game ends
listener.stop()
````

````python
import random
import os
import time

def rollDice(side):
    result = random.randint(1, side)
    return result

def health():
    healthStat = ((rollDice(6) * rollDice(12)) / 2) + 10
    return healthStat

def strength():
    strengthStat = ((rollDice(6) * rollDice(8)) / 2) + 12
    return strengthStat

def create_character():
    while True:
        name = input("Name your Legend:\n")
        if name.strip():
            break
        print("Name cannot be empty. Please enter a valid name.")
    
    character_type = input("Character Type (Human, Elf, Wizard, Orc):\n")
    health_stat = health()
    strength_stat = strength()
    
    return {
        "name": name,
        "type": character_type,
        "health": health_stat,
        "strength": strength_stat
    }

def battle(character1, character2):
    round_counter = 0
    while character1["health"] > 0 and character2["health"] > 0:
        os.system('cls' if os.name == 'nt' else 'clear')
        print("‚öîÔ∏è BATTLE TIME ‚öîÔ∏è")
        print()
        print("The battle begins!")
        
        roll1 = rollDice(6)
        roll2 = rollDice(6)
        
        if roll1 > roll2:
            winner = character1
            loser = character2
        elif roll2 > roll1:
            winner = character2
            loser = character1
        else:
            print(f"Their swords clash and they draw round {round_counter + 1}")
            print()
            print(f"{character1['name']}")
            print(f"HEALTH: {character1['health']}")
            print()
            print(f"{character2['name']}")
            print(f"HEALTH: {character2['health']}")
            print()
            print("And they're both standing for the next round")
            time.sleep(2)
            continue
        
        damage = abs(winner["strength"] - loser["strength"]) + 1
        loser["health"] -= damage
        
        round_counter += 1
        print(f"{winner['name']} wins round {round_counter}")
        print()
        print(f"{character1['name']}")
        print(f"HEALTH: {character1['health']}")
        print()
        print(f"{character2['name']}")
        print(f"HEALTH: {character2['health']}")
        print()
        print("And they're both standing for the next round")
        time.sleep(2)
    
    os.system('cls' if os.name == 'nt' else 'clear')
    print("‚öîÔ∏è BATTLE TIME ‚öîÔ∏è")
    print()
    if character1["health"] <= 0:
        print(f"{character1['name']} has died!")
        print(f"{character2['name']} has won in {round_counter} rounds")
    else:
        print(f"{character2['name']} has died!")
        print(f"{character1['name']} has won in {round_counter} rounds")

while True:
    print("‚öîÔ∏è CHARACTER BUILDER ‚öîÔ∏è")
    print()
    
    print("Who is the first character?")
    character1 = create_character()
    
    print()
    print("Who are they battling?")
    character2 = create_character()
    
    battle(character1, character2)
    
    again = input("Again? (yes/no):\n")
    if again.lower() == "no":
        break
    
    time.sleep(1)
    os.system('cls' if os.name == 'nt' else 'clear')
````

```
myString = "Hello there my friend."

```

```
myString = "Hello there my friend."
print(myString[6:11])

# This code outputs 'there'.
```

3. Leaving the first index blank defaults to 'start from index 0'.

```
myString = "Hello there my friend."
print(myString[:11])

# This code outputs 'Hello there'.
```

4. Leaving the last index blank defaults to 'go to the end'.

```
myString = "Hello there my friend."
print(myString[12:])
# This code outputs 'my friend.'.
```

````python
import random
import os
import time

listOfWords = ["apple", "orange", "grapes", "pear"]
letterPicked = []
lives = 6

word = random.choice(listOfWords)

# Display the number of letters in the word from the beginning
print("The word has", len(word), "letters.")
print(" ".join("_" * len(word)))

while True:
    time.sleep(1)
    os.system('cls' if os.name == 'nt' else 'clear')
    letter = input("Choose a letter: ").lower()
    
    if letter in letterPicked:
        print("You've tried that before")
        continue
    
    letterPicked.append(letter)
    
    if letter in word:
        print("You found a letter")
    else:
        print("Nope, not in there")
        lives -= 1
    
    allLetters = True
    print()
    for i in word:
        if i in letterPicked:
            print(i, end=" ")
        else:
            print("_", end=" ")
            allLetters = False
    print()
    
    # Display the letters that have already been used
    print("Letters used:", " ".join(letterPicked))
    
    if allLetters:
        print(f"You won with {lives} lives left!")
        break

    if lives <= 0:
        print(f"You ran out of lives! The answer was {word}")
        break
    else:
        print(f"Only {lives} lives left")
````
