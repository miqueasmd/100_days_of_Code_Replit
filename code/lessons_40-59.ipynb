# 🌟 Day 41 Challenge 🌟

### Task:
Create a dictionary that stores the following information about a website:
1. Name
2. URL
3. Description
4. A star rating (out of 5)

### Steps:
1. Use a loop to output the names of the keys.
2. Ask the user to type in the details and store the input in the dictionary.
3. Finally, output the whole dictionary (keys and values).

### Example:

🌟 **Website Rating** 🌟

```plaintext
Input the website name: Replit
Input the URL: replit.com
Input your a description: An awesome online IDE.
Input the a star rating out of 5: *****

name: Replit, URL: replit.com, description: An awesome online IDE, rating: *****
```

### Hints:
1. When creating your dictionary, you will need to use ```example = { "MyValue": none}``` to show a key name and no value.
2. Use a loop to ```print``` the entire dictionary.
3. Make sure you include both variables (name and value) in your loop and your ```print``` statement.

````python
website = {"name": None, "url": None, "desc": None, "rating": None}

# Function to get and validate rating input
def get_rating():
    while True:
        try:
            rating = float(input("rating (0-5): ").strip())
            if 0 <= rating <= 5:
                return rating
            else:
                print("Please enter a rating between 0 and 5.")
        except ValueError:
            print("Invalid input. Please enter a number.")

# Get user input for each field
for key in website.keys():
    if key == "rating":
        website[key] = get_rating()
    else:
        website[key] = input(f"{key}: ").strip()

# Convert rating to stars
def rating_to_stars(rating):
    return '★' * int(rating) + '☆' * (5 - int(rating))

# Print the website details
print("\nWebsite Details:")
for key, value in website.items():
    if key == "rating":
        print(f"{key.capitalize()}: {rating_to_stars(value)} ({value})")
    else:
        print(f"{key.capitalize()}: {value}")
````

# 🌟 Day 44 Challenge 🌟

It's time for more bingo! You can reuse your code from Day 43, but this time add the following features:

### Steps:
1. Repeatedly ask the user what number comes up next.
2. Check the bingo card to see if the number picked matches one on the card.
3. If the bingo card is all **X's**, then the user has won.

### Example:

🌟 **Bingo Game** 🌟

```plaintext
2    |  4   |  X  
X    |  BG  |  46 
61   |  68  |  X  

Next Number: __

X    |  X   |  X  
X    |  BG  |  X  
X    |  68  |  X  

Next Number: 68  
You have won
```

### Hints:
1. Create a subroutine called ```createCard``` to clean up some of the code from Day 43.
2. Use a variable and loop to store how many x's there are on the card. Add one every time a number is replaced.
3. Check the variable every time to see if it has reached the magic winning number.

````python
import random
import os
import time

bingo = []

def ran():
    number = random.randint(1, 90)
    return number

def prettyPrint():
    for row in bingo:
        print(" | ".join(f"{str(item):^5}" for item in row))
        print("-" * 35)  # Add a separator line for better readability

def createCard():
    global bingo
    numbers = set()
    while len(numbers) < 24:  # We need 24 unique numbers (excluding the center "BINGO" cell)
        numbers.add(ran())
    
    numbers = sorted(numbers)
    
    bingo = [
        [numbers[0], numbers[1], numbers[2], numbers[3], numbers[4]],
        [numbers[5], numbers[6], numbers[7], numbers[8], numbers[9]],
        [numbers[10], numbers[11], "BINGO", numbers[12], numbers[13]],
        [numbers[14], numbers[15], numbers[16], numbers[17], numbers[18]],
        [numbers[19], numbers[20], numbers[21], numbers[22], numbers[23]]
    ]

createCard()
while True:
    os.system('cls' if os.name == 'nt' else 'clear')
    prettyPrint()
    num = input("Next Number (or 0 to exit): ")
    
    if num == "0":
        print("Exiting the game.")
        break
    
    try:
        num = int(num)
    except ValueError:
        print("Invalid input. Please enter a number.")
        time.sleep(1)
        continue
    
    for row in range(5):
        for item in range(5):
            if bingo[row][item] == num:
                bingo[row][item] = "X"

    exes = 0
    for row in bingo:
        for item in row:
            if item == "X":
                exes += 1

    if exes == 24:  # 24 because the center "BINGO" cell is not counted
        print("You have won")
        break

    time.sleep(1)
    print("\n")  # Add a blank line between each print of the bingo card
````

# 🌟 Day 45 Challenge 🌟

Made it? Good! Let’s get cracking.

### Task:
Your system should:
1. Have a menu that asks if you want to add, view, move, or edit a 'to do'.
2. If you choose 'add', then the system should:
   - Prompt you to input what the 'to do' is, when it is due by, and the priority (high, medium, or low).
   - Add the 'to do' to the list.
3. 'View' should give two options:
   - **View all**: Shows all 'to dos' with a pretty print.
   - **View priority**: Allows you to search for high, medium, or low priority tasks and only see matching tasks.
4. 'Edit' allows you to change any of the information within one of the 'to dos'.
5. 'Remove' lets you completely remove a 'to do' when it is 'to done'.

### Example:

🌟 **Life Organizer** 🌟

```plaintext
Welcome to the life organizer. Do you want to add, view, edit or remove a to do? > Add
```

### Hints:
1. Use a separate subroutine for add, view, edit, and remove.
2. Clear the console before viewing a new entry.
3. Use a ```while True``` loop to call the subroutines and display the menu.

````python
import os
import time

# Initialize the to do list
todo = []

# Function to clear the screen
def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

# Function to add a new task to the to-do list
def add():
    clear_screen()
    name = input("Name > ").strip()
    date = input("Due Date > ").strip()
    priority = input("Priority (Low, Medium, High) > ").strip().capitalize()
    row = [name, date, priority]
    todo.append(row)
    print("Added")
    time.sleep(1)
    printList()

# Function to view tasks in the to-do list
def view():
    clear_screen()
    if not todo:
        print("Your to do list is empty.")
    else:
        options = input("1: All\n2: By Priority\n> ").strip()
        if options == "1":
            # View all tasks
            for row in todo:
                print(" | ".join(row))
            print()
        elif options == "2":
            # View tasks by priority
            priority = input("What priority? (Low, Medium, High) > ").strip().capitalize()
            found = False
            for row in todo:
                if priority in row:
                    print(" | ".join(row))
                    found = True
            if not found:
                print(f"No tasks with {priority} priority.")
            print()
        else:
            print("Invalid option.")
    time.sleep(1)

# Function to edit an existing task in the to-do list
def edit():
    clear_screen()
    find = input("Name of todo to edit > ").strip()
    found = False
    for row in todo:
        if find in row:
            found = True
            todo.remove(row)
            name = input("Name > ").strip()
            date = input("Due Date > ").strip()
            priority = input("Priority (Low, Medium, High) > ").strip().capitalize()
            row = [name, date, priority]
            todo.append(row)
            print("Updated")
            break
    if not found:
        print("Couldn't find that")
    time.sleep(1)

# Function to remove a task from the to-do list
def remove():
    clear_screen()
    find = input("Name of todo to remove > ").strip()
    found = False
    for row in todo:
        if find in row:
            todo.remove(row)
            found = True
            print("Removed")
            break
    if not found:
        print("Couldn't find that")
    if not todo:
        print("Your to do list is now empty.")
    time.sleep(1)

# Function to print the entire to-do list
def printList():
    print("\nTo Do List:")
    if not todo:
        print("Your to do list is empty.")
    else:
        for index, item in enumerate(todo, start=1):
            print(f"{index}. {' | '.join(item)}")
    print()

# Main loop to prompt the user for actions
while True:
    clear_screen()
    menu = input("1: View\n2: Add\n3: Edit\n4: Remove\n5: Exit\n> ").strip().lower()
    
    if menu == "1" or menu == "view":
        view()
    elif menu == "2" or menu == "add":
        add()
    elif menu == "3" or menu == "edit":
        edit()
    elif menu == "4" or menu == "remove":
        remove()
    elif menu == "5" or menu == "exit":
        print("Exiting ToDoList Manager.")
        break
    else:
        print("Invalid action. Please choose '1', '2', '3', '4', or '5'.")
        time.sleep(1)
````

### Hints:
If you're adding your own cards dynamically, try using the ```random.choice()``` function to 'draw' two cards from the deck.

````python
import os
import time
import random

# Initialize the trumps dictionary with characters and their stats
trumps = {
    "David": {"Intelligence": 178, "Speed": 4, "Guile": 80, "Baldness Level": 99},
    "Mr Spock": {"Intelligence": 200, "Speed": 50, "Guile": 50, "Baldness Level": 0},
    "Monica from Friends": {"Intelligence": 150, "Speed": 50, "Guile": 50, "Baldness Level": 0},
    "Professor X": {"Intelligence": 250, "Speed": 1, "Guile": 200, "Baldness Level": 101}
}

# Function to clear the screen
def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

# Main game loop
while True:
    clear_screen()
    print("TOP TRUMPS")
    print("\nCharacters\n")
    
    # Display available characters
    for key in trumps:
        print(key)
    
    # Get user character choice
    user = input(f"\nPick your character ({', '.join(trumps.keys())})\n> ").strip()
    
    # Ensure the user picks a valid character
    if user not in trumps:
        print("Invalid character. Please choose from the list.")
        time.sleep(2)
        continue
    
    # Computer randomly picks a character
    comp = random.choice(list(trumps.keys()))
    print("Computer has picked", comp)
    
    # Get user stat choice
    stats = list(trumps[user].keys())
    answer = input(f"\nChoose your stat ({', '.join(stats)})\n> ").strip().title()
    
    # Ensure the user picks a valid stat
    if answer not in stats:
        print("Invalid stat. Please choose from the list.")
        time.sleep(2)
        continue
    
    # Display the chosen stats for both characters
    print(f"\n{user}: {trumps[user][answer]}")
    print(f"{comp}: {trumps[comp][answer]}")
    
    # Determine and display the winner
    if trumps[user][answer] > trumps[comp][answer]:
        print(f"\n{user} wins!")
    elif trumps[user][answer] < trumps[comp][answer]:
        print(f"\n{comp} wins!")
    else:
        print("\nIt's a draw!")
    
    # Ask the user if they want to play again or exit
    play_again = input("\nDo you want to play again? (y/n) > ").strip().lower()
    if play_again != 'y':
        print("Exiting the game. Goodbye!")
        break
    
    # Pause before the next round
    time.sleep(2)
    clear_screen()
````

4. Each new entry score should go on a new line as `initials score`. Then start a new line for the next entry.

### Example:


### Hints:

1. Not much here that you can't get from the examples.

````python
import os
import time
import random
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()
data_path = os.getenv('DATA_PATH')

# Ensure the DATA_PATH environment variable is set
if not data_path:
    print("Error: DATA_PATH environment variable is not set.")
    exit(1)

# Ensure the data_path ends with a slash
if not data_path.endswith('/'):
    data_path += '/'

# File path for the to-do list
file_path = data_path + "to_do.txt"

# Initialize the to-do list
todo = []

# Boolean variable to check if the file exists
fileExists = True

# Load the to-do list from the file
def load_todo():
    global todo, fileExists
    try:
        with open(file_path, "r") as f:
            todo = eval(f.read())
    except FileNotFoundError:
        todo = []
        fileExists = False

# Save the to-do list to the file
def save_todo():
    # Create a backup folder if it doesn't exist
    backup_folder = data_path + "backups/"
    if not os.path.exists(backup_folder):
        os.mkdir(backup_folder)

    # Create a random filename for the backup
    backup_filename = f"backup{random.randint(1, 10)}.txt"
    backup_filepath = backup_folder + backup_filename

    # Save a copy of the data to the backup file
    with open(backup_filepath, "w") as backup_file:
        backup_file.write(str(todo))

    # Save the to-do list to the main file
    with open(file_path, "w") as f:
        f.write(str(todo))

# Function to clear the screen
def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

# Function to add a new task to the to-do list
def add():
    clear_screen()
    name = input("Name > ").strip()
    date = input("Due Date > ").strip()
    priority = input("Priority (Low, Medium, High) > ").strip().capitalize()
    row = [name, date, priority]
    todo.append(row)
    print("Added")
    save_todo()
    time.sleep(1)
    printList()

# Function to view tasks in the to-do list
def view():
    clear_screen()
    if not todo:
        print("Your to do list is empty.")
    else:
        options = input("1: All\n2: By Priority\n> ").strip()
        if options == "1":
            # View all tasks
            for row in todo:
                print(" | ".join(row))
            print()
        elif options == "2":
            # View tasks by priority
            priority = input("What priority? (Low, Medium, High) > ").strip().capitalize()
            found = False
            for row in todo:
                if priority in row:
                    print(" | ".join(row))
                    found = True
            if not found:
                print(f"No tasks with {priority} priority.")
            print()
        else:
            print("Invalid option.")
    time.sleep(1)

# Function to edit an existing task in the to-do list
def edit():
    clear_screen()
    find = input("Name of todo to edit > ").strip()
    found = False
    for row in todo:
        if find in row:
            found = True
            todo.remove(row)
            name = input("Name > ").strip()
            date = input("Due Date > ").strip()
            priority = input("Priority (Low, Medium, High) > ").strip().capitalize()
            row = [name, date, priority]
            todo.append(row)
            print("Updated")
            save_todo()
            break
    if not found:
        print("Couldn't find that")
    time.sleep(1)

# Function to remove a task from the to-do list
def remove():
    clear_screen()
    find = input("Name of todo to remove > ").strip()
    found = False
    for row in todo:
        if find in row:
            todo.remove(row)
            found = True
            print("Removed")
            save_todo()
            break
    if not found:
        print("Couldn't find that")
    if not todo:
        print("Your to do list is now empty.")
    time.sleep(1)

# Function to print the entire to-do list
def printList():
    print("\nTo Do List:")
    if not todo:
        print("Your to do list is empty.")
    else:
        for index, item in enumerate(todo, start=1):
            print(f"{index}. {' | '.join(item)}")
    print()

# Load the to-do list when the program starts
load_todo()

# Main loop to prompt the user for actions
while True:
    clear_screen()
    menu = input("1: View\n2: Add\n3: Edit\n4: Remove\n5: Exit\n> ").strip().lower()
    
    if menu == "1" or menu == "view":
        view()
    elif menu == "2" or menu == "add":
        add()
    elif menu == "3" or menu == "edit":
        edit()
    elif menu == "4" or menu == "remove":
        remove()
    elif menu == "5" or menu == "exit":
        print("Exiting ToDoList Manager.")
        break
    else:
        print("Invalid action. Please choose '1', '2', '3', '4', or '5'.")
        time.sleep(1)
````

`1 * 2 * 3 * 4 * 5 = 120`

   - **Multiplies** that by factorial of itself minus one.

# 🌟 Day 59 Challenge 🌟

You're going to write a program that checks a string to see if it is a palindrome.

Yes, we **know** that Python has the built-in function `string.reverse()` that you could use.

Zero points for that today; we want you to think hard and utilize your skills in:
- recursion
- string slicing
- looping


### Task:
Your program should:
1. Prompt the user to input a word.
2. Analyze the word to see if it is a palindrome.
3. Output a relevant yes/no message.


### Example:

🌟 **Palindrome Checker** 🌟

```plaintext
Input a word > Racecar

Racecar is a palindrome. Yay!
```

### Hints:
This is a tough one, so I've given you some hints about the algorithmic thinking needed for a problem like this:
1. Don't forget to standardize the case on input.
2. Think about which characters in a word are compared first. Check if they are the same.
3. If they've been compared and are the same, remove them and repeat the process with the shorter string.
4. Keep going until you're down to a string of length 1 or 0 (depending on whether the original word had an odd/even number of characters. If you get to this point, then you've got a palindrome.

````python
def palindrome(word):
    """
    Check if a word is a palindrome.
    
    A palindrome is a word that reads the same backward as forward.
    
    Parameters:
    word (str): The word to check.
    
    Returns:
    bool: True if the word is a palindrome, False otherwise.
    """
    # Base case: if the word has 1 or 0 characters, it is a palindrome
    if len(word) <= 1:
        return True
    
    # Check if the first and last characters are the same
    if word[0] != word[-1]:
        return False
    
    # Recursive case: check the substring excluding the first and last characters
    return palindrome(word[1:-1])

# Main function to prompt the user for input and check if it is a palindrome
def main():
    try:
        word = input("Introduce the word that you want to check: ").strip()
        if not word.isalpha():
            raise ValueError("Input must be a single word containing only alphabetic characters.")
        if palindrome(word):
            print(f"'{word}' is a palindrome.")
        else:
            print(f"'{word}' is not a palindrome.")
    except ValueError as e:
        print(e)

# Run the main function
if __name__ == "__main__":
    main()
````
